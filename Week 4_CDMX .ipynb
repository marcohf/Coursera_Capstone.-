## Let's install Geocoder and make the process to get the latitudes and longitudes

```python
! pip install geocoder
```

```python
import geocoder

# Defining a function to get coordinates

def get_latlng(neighborhood):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, CDMX,MX'.format(neighborhood))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

##  Create a temporary dataframe to populate the coordinates into Latitude and Longitude

```python
# Create temporary dataframe to populate the coordinates into Latitude and Longitude
df_coords = pd.DataFrame(coords, columns=['Latitude', 'Longitude'])
# Merge the coordinates into the original dataframe
dfCDMX['Latitude'] = df_coords['Latitude']
dfCDMX['Longitude'] = df_coords['Longitude']
print( dfCDMX.shape)
dfCDMX.head(5)
```
